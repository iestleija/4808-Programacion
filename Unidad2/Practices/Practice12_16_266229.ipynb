## Práctica 12: Carga de Datos

Al final del archivo contiene una línea que dice "Esta es una línea que no deberías cargar". Esta línea es irrelevante por lo tanto al cargar el CSV a un dataframe se debería de ignorar.

Carga el archivo a un dataframe usando el código necesario para omitir la última fila.
Usa df.tail(3) para demostrar que no has cargado la última línea
Imprime la cantidad de filas y columnas del dataframe

```python

# rss stands for real state sales
rss = pd.read_csv('/content/drive/MyDrive/4808-programacion/U02/A12-16/miaad-nyc-r-s.csv', skipfooter=1, engine='python')
rss.tail(3)
```

```python
# save the original data dictionary
dd_original = pd.DataFrame(
    {"Nombre de columna": rss.columns,
     "Tipo de dato": [str(rss[col].dtype) for col in rss.columns]}
);
```

## Práctica 13: Cambiar nombre de columna

Imprime los nombres de la columnas del archivo
Asegúrate que el nombre de la primera columna debería ser INDEX MIAAD. Si esto no es así, entonces cambia el nombre de la columna y vuelve a imprimir los nombres para mostrar el cambio realizado. Después del cambio debería aparecer INDEX MIAAD

```python
# original columns names
print_bold("Nombres de columnas originales.")
print(rss.columns);

# making the required renaming
rss = rss.rename(columns={'Unnamed: 0': 'INDEX MIAAD'})

# updated cloumns names
print_bold("Nombres de columnas actualizados.")
print(rss.columns);
```

## Práctica 14: Normalización de los nombres de la columna.

Normaliza los nombres de las columnas, los cuales deberán estar en minúsculas. Asimismo, reemplaza los espacios con guion bajo
Imprime los nombres de las columnas para demostrar que has realizado la normalización.

```python
# columns names converted to lower case and to snake case using underscores instead of spaces.
rss.columns = rss.columns.str.lower().str.replace(' ', '_')
print_bold("The updated columns names in lower and snake case are:\n")
print(rss.columns);
```

## Práctica 15: Errores en los Tipos de Datos

Imprime los tipos de datos de cada columna.
Asegúrate que las columnas de SALE PRICE(sale_price), LAND SQUARE FEET (land_square_feet), GROSS SQUARE FEET (gross_square_feet) sean de tipo flotante. Escribe qué tipo de dato tienen asignado. Si no son correctos, realiza el cambio.
Imprime de nuevo los tipos de datos para demostrar que has arreglado el problema.

```python
# required columns to convert their data type to float
cols_to_fix = ['sale_price', 'land_square_feet', 'gross_square_feet'];

#current data types
print_bold("Showing only the required columns")
print_bold("The original data types are:");
print(rss[cols_to_fix].dtypes);

#applying the change
rss[cols_to_fix] = rss[cols_to_fix].apply(pd.to_numeric, errors='coerce');

# fixed data types for the required columns
print_bold("The fixed data types are:");
print(rss[cols_to_fix].dtypes);
```

## Práctica 16: Genera el diccionario de datos del dataframe preprocesado. Imprime el contenido de dicho diccionario
Habiendo generado el dicionario de datos antes de realizar cambios al dataframe, se muestran a continuación el original y el que resultó despues de los cambios.

```python
dd_corregido = pd.DataFrame(
    {"Nombre de columna": rss.columns,
     "Tipo de dato": [str(rss[col].dtype) for col in rss.columns]}
);

display(Markdown("**El diccionario de datos original es:**"))
print(dd_original)

display(Markdown("**El diccionario de datos corregido es:**"))
print(dd_corregido)
```
