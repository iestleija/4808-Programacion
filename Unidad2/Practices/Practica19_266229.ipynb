## Ejercicio 01 Duplicidad Parcial Una columna: Contar y Filtrar los registros duplicados en Duration


### 1. Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.


```python
dirty_df = pd.read_csv(file_path)
print_bold("Carga de dataframe original para ejercicio 1")
dirty_df.head()
```

### 2. Cuenta la cantidad de registros duplicados utilizando solo la información de columna Duration e imprime
3. Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.


```python
# calculating exercise 1 duplicates in a dedicated copy
dirty_df_dup1 = dirty_df.copy()
print_bold("Datos duplicados en la columna Duration")

# Add a flag to mark the duplicate rows.
dirty_df_dup1["Duration Duplicate"] = dirty_df.duplicated(subset=["Duration"],keep=False)

# Show the counting
dirty_df_dup1["Duration Duplicate"].value_counts()
```

### 4. Filtra los datos y obten los registros sin duplicados


```python
print_bold("Filas sin duplicados en columna Duration")
dirty_df_dup1[dirty_df_dup1["Duration Duplicate"] == False]
```

## Ejercicio 02 Duplicidad Parcial Dos columnas: Contar y Filtrar los registros duplicados en Pulse y MaxPulse



### 1. Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.


```python
dirty_df = pd.read_csv(file_path)
print_bold("Carga de dataframe original para ejercicio  2")
dirty_df.head()

```

### 2. Cuenta la cantidad de registros duplicados utilizando la información de las columnas Pulse y Maxpulse
3. Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.
---
En esta primera instancia se marcan los duplicados sin recuperación con `keep=False`

```python
# calculating exercise 2 duplicates in a dedicated copy
dirty_df_dup2 = dirty_df.copy()
print_bold("Datos duplicados usando las columnas Pulse y Maxpulse, con keep=False")
dirty_df_dup2["Pulse and Maxpulse Duplicate"] = dirty_df_dup2.duplicated(subset=["Pulse", "Maxpulse"],keep=False)
dirty_df_dup2["Pulse and Maxpulse Duplicate"].value_counts()
```

### 4. Filtra los datos y obten los registros cuando keep = 'last'
---
Nótese la diferencia: cuando `keep='last'`, se recupera uno de los duplicados, en este caso la última aparición.

```python
dirty_df_dup2["Pulse and Maxpulse Duplicate"] = dirty_df_dup2.duplicated(subset=["Pulse", "Maxpulse"],keep='last')
print_bold("Filas duplicados usando las columnas Pulse y Maxpulse con keep='last'")
dirty_df_dup2[dirty_df_dup2["Pulse and Maxpulse Duplicate"] == True]
```

## Ejercicio 03 Duplicidad Total o Exacta: Contar y Filtrar los registros duplicados


1. Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.

```python
dirty_df = pd.read_csv(file_path)
print_bold("Carga de dataframe original para ejercicio  3")
dirty_df.head()
```

2. Cuenta la cantidad de registros duplicados exactos (todas las columnas)


```python
# calculating exercise 3 duplicates in a dedicated copy
dirty_df_dup3 = dirty_df.copy()
dirty_df_dup3["Duplicate"] = dirty_df_dup3.duplicated(keep=False)
print_bold("Conteo de filas completas duplicadas en el dataframe original")
dirty_df_dup3["Duplicate"].value_counts()
```

### 3. Utiliza sum() e imprime la cantidad de registros con duplicidad exacta

```python
total_duplicates = dirty_df_dup3.duplicated(keep=False).sum()
print_bold(f"En total hay {total_duplicates} filas completas duplicadas.")
```

### 4. Filtra los datos y obten los registros cuando keep = 'last'

```python
dirty_row_duplicates= dirty_df_dup3[dirty_df_dup3.duplicated(keep='last')]
print_bold("Filas duplicadas dejando una (la última) de cada conjunto de filas duplicadas")
print(dirty_row_duplicates)
```

## Ejercicio 04: Cálculo de la media con duplicados y sin duplicados



### 1. Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.



```python
dirty_df = pd.read_csv(file_path)
print_bold("Carga de dataframe original para ejercicio 4")
dirty_df.head()
```

### 2. Calcula e imprime la media de las columnas Pulse y MaxPulse


```python
print_bold("Media aritmética de las columnas Pulse y Maxpulse con datos originales")
print(dirty_df[["Pulse", "Maxpulse"]].mean())
print_bold("Forma inicial de dataframe original")
dirty_df.shape
```

### 3. Identifica los valores duplicados en las columnas Pulse y MaxPulse (Duplicidad Parcial) y elimina los registro repetidos. Quédate con las últimas ocurrencias

```python
dirty_df_dup4 = dirty_df.drop_duplicates(subset=["Pulse", "Maxpulse"], keep="last")
print_bold("Forma del dataframe sin duplicados para columnas Pulse y Maxpulse")
print_bold("Se eliminaron 3 filas")
dirty_df_dup4.shape
```

### 4. Utiliza el conjunto anterior sin datos duplicados (recuerda que solo quedaron las últimas ocurrencias de los datos) y calcula la media de las columnas Pulse y MaxPulse


```python
print_bold("Valores del dataframe ORIGINAL en columnas Pulse y Maxpulse")
print(dirty_df[["Pulse", "Maxpulse"]].mean())
print_bold("Valores del dataframe SIN DUPLICADOS en columnas Pulse y Maxpulse")
print(dirty_df_dup4[["Pulse", "Maxpulse"]].mean())
```
