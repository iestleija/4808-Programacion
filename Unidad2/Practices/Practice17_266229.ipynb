```python
#Cargar la biblioteca
import pandas as pd

#Leemos el CSV cat_err01.csv
df = pd.read_csv('/content/drive/MyDrive/4808-programacion/U02/P17/nls97.csv',index_col=0)
df.info()
```

###Funciones utilitarias
Usaremos estas funciones en las siguientes secciones

```python
def print_bold(msg):
    print("\033[1m" + msg + "\033[0m")

# Recibe una columna y un diccionario de remplazos para modificar los valores
# de una columna del unico dataframe en este documento.
def corregir_con_replace(col, diccionario_correccion):
  # Categorías de columna originales
  print_bold(f"Valores únicos en {col}:");
  print(df[col].unique())
  print_bold(f"Frecuencia original de valores únicos en {col}:");
  print(f"{df[col].value_counts().sort_index()}")

  # Aplicar diccionario de correccion.
  df[col] = df[col].replace(diccionario_correccion)

  # Categorías de columna corregidas
  print_bold(f"Valores únicos en {col}:");
  print(df[col].unique())
  print_bold(f"Frecuencia corregida de valores únicos en {col}:");
  print(df[col].value_counts().sort_index())
```

## Ejercicio 01: Errores Tipográficos y Valores Inconsistentes: maritalstatus

1. Imprime de la columna 'maritalstatus' los valores únicos.
2. Imprime la frecuencia de cada valor único
3. Cambia el valor 'Never-Married' a 'Single'
4. Imprime la frecuencia de cada valor único.

```python
# Se usa la función previamente definida para ejecutar los pasos requeridos.
# Diccionario de correccion de errores tipográficos e inconsistencias
correccion_maritalstatus = {
    'Never-married': 'Single',
}

corregir_con_replace('maritalstatus', correccion_maritalstatus)
```

## Ejercicio 02: Errores Tipográficos y Valores Inconsistentes: highestdegree.

1. Imprime de la columna 'highestdegrees' los valores únicos.
2. Imprime la frecuencia de cada valor único
2. Observa que cada valor categórico empieza con 0. Modifica las categorías para que todas estén en minúsculas y ninguna empiece con un número. Por ejemplo, si dice "2. High School", debería ser 'high school.
3. Los cambios se realizarán en la misma columna 'highestdegrees'.
4. Imprime los nuevos valores únicos.

```python
col = 'highestdegree'
# 1. Categorías de columna originales
print_bold(f"Valores únicos en {col}:");
print(df[col].unique())
# 2.  Categorías de columna originales
print_bold(f"Frecuencia original de valores únicos en {col}:");
print(f"{df[col].value_counts().sort_index()}")

# 3 y 4 . Aplicación de los cambios requeridos en un solo paso.
# - Se aplica la funcion regex para eliminar el prefijo no deseado
# - Se eliminan los espacios excedentes
# - Se convierten todas las etiquetad de categoría a minúsculas
df[col] = df[col].str.replace(r'^\s*[0-9]+\.\s*', '', regex=True).str.strip().str.lower()

# 5. Categorías de columna corregidas
print_bold(f"Valores únicos en {col}:");
print(df[col].unique())
print_bold(f"Frecuencia corregida de valores únicos en {col}:");
print(df[col].value_counts().sort_index())
```

## Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas

1. De la columna 'maritalstatus' une las categorías Widowed y 'Single' en una categoría más general llamada 'Single/No Partner'.
Crea una nueva columna, 'marital_grouped' donde aparezcan las nuevas categorías de maritalstatus.

2. Imprime los valores únicos de la nueva columna 'marital_grouped'.

3. Imprime la frecuencia de los valores únicos de la nueva columna 'marital_grouped'

```python
def merge_single_and_widowed(marital_status):
  if marital_status == 'Widowed' or marital_status == 'Single':
    return 'Single/No Partner'
  else:
    return marital_status

col = 'maritalstatus'
# Categorías de columna originales
print_bold(f"Valores únicos en {col}:");
print(df[col].unique())
print_bold(f"Frecuencia original de valores únicos en {col}:");
print(f"{df[col].value_counts().sort_index()}")

# 1. New column for marital status merging single and widowed categories.

df['marital_grouped'] = df['maritalstatus'].apply(merge_single_and_widowed)

col2 = 'marital_grouped'
# 2. Categorías de columna originales
print_bold(f"Valores únicos en {col2}:");
print(df[col2].unique())

# 3. Categorías de columna originales
print_bold(f"Frecuencia original de valores únicos en {col2}:");
print(f"{df[col2].value_counts().sort_index()}")
```
