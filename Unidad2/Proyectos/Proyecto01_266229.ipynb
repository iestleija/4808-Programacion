# Sección de Respuestas

**Preparación de la Correlación de Datos Socio-demográficos**

```python
from numpy._core.defchararray import index

coexisting_factors = ["age", "sex", "anaemia", "diabetes", "high_blood_pressure", "smoking", "DEATH_EVENT"]
correlation_matrix = heart_failure[coexisting_factors].corr(numeric_only=True)
correlation_matrix.rename(
    columns={'age':'Edad','sex':'Género', 'anaemia':'Anemia','diabetes':'Diabetes','high_blood_pressure':'Presión alta','smoking':'Fumador','DEATH_EVENT':'Muerte'},
    index={'age':'Edad','sex':'Género', 'anaemia':'Anemia','diabetes':'Diabetes','high_blood_pressure':'Presión alta','smoking':'Fumador','DEATH_EVENT':'Muerte'},
    inplace=True)
print(correlation_matrix)
```

### Correlación de las características del perfil con la supervivencia.


```python

#Heatmap graph definition
sns.heatmap(correlation_matrix, cmap="Blues", annot=True) #cmaps ["coolwarm", "RdBu", "vlag","Blues", "rocket", "viridis","YlGnBu"]

# customization
plt.title("Correlación de Factores Individuales: Después de Falla Cardíaca")
plt.show()
```

##Respuesta a Pregunta 1

De la vista de la correlación entre los distintos factores del perfil individual con el dato objetivo, podemos decir que no se encuentra relación importante de ninguno de ellos con la mortalidad, siendo la correlación mas prominente la edad de cada paciente.

##Respuesta a Pregunta 5
¿ Que impacto tuvo la diabetes en la mortalidad después de la falla cardíaca ?

Confirmando la información del mapa de calor de la Pregunta 1, la comparativa muestra que la diabetes no aportó a las muertes despues de una falla cardíaca.

```python
# Crear la figura y los subplots (2 filas, 2 columnas)
fig, axs = plt.subplots(1, 2, figsize=(10, 3));
plt.suptitle(r'Impacto de la Diabetes en la Fatalidad', fontsize=16)

death_counts_with_diabetes = heart_failure[heart_failure['diabetes'] == 1]['DEATH_EVENT'].value_counts()
death_counts_without_diabetes = heart_failure[heart_failure['diabetes'] == 0]['DEATH_EVENT'].value_counts()
mis_etiquetas = ['\u2611 Sobrevivió', '\u2620 Murió']
death_counts_with_diabetes.index = mis_etiquetas
death_counts_without_diabetes.index = mis_etiquetas

death_counts_with_diabetes.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'], ax=axs[0]);
axs[0].set_xlabel("Con diabetes", fontsize=12, labelpad=20);
axs[0].set_ylabel(""); # Elimina la etiqueta del eje Y para que se vea más limpio

death_counts_without_diabetes.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'], ax=axs[1]);
axs[1].set_xlabel("Sin diabetes", fontsize=12, labelpad=20);
axs[1].set_ylabel(""); # Elimina la etiqueta del eje Y para que se vea más limpio

plt.tight_layout();
```
