En el dataset en cuestión no tenemos variables temporales, porque la variable `time` es solo un indicador de cuantos días se dió seguimiento.

La única conversión recomendada es hacia variables categóricas para las variables identificadas previamente.

La conversión a tipo categórico puede optimizar el uso de memoria para columnas con pocos valores únicos y puede ser importante para ciertos tipos de análisis y modelos de aprendizaje automático.


```python
# Preprocesamiento de datos
# Revisión y corrección de: Tipos de datos, Variables categóricas y variables temporales.
# Solo aplicó para variables categóricas.

categoric_variables= ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking', 'death_event']

print_bold("Ahorro de memoria usando el tipo de datos category\n")

# Revisar el uso de memoria en dataset original
print(f"El uso de memoria original es: {heart_failure.memory_usage(deep=True).sum()} bytes");
preprocessed_heart_failure = heart_failure.copy()

# Convertir la columna 'anaemia' a tipo categórico
for variable in categoric_variables:
  preprocessed_heart_failure[variable] = preprocessed_heart_failure[variable].astype('category')

print(f"El uso de memoria con valores categóricos  es: {preprocessed_heart_failure.memory_usage(deep=True).sum()} bytes");
#display(heart_failure['anaemia'].value_counts())
```

```python
# Análisis Multivariado
# Matriz de Correlación para Datos Bio-químicos

from numpy._core.defchararray import index
bio_chem_renames = {'creatinine_phosphokinase': 'Creatinina Fosfoquinasa','ejection_fraction':'Fracción de Eyección','platelets':'Plaquetas','serum_creatinine':'Creatinina en Suero','serum_sodium':'Sodio en Suero','death_event':'Muerte'}

bio_chem_corr_matrix = preprocessed_heart_failure[bio_chemical_vars + ['death_event'] ].corr()
bio_chem_corr_matrix.rename(
    columns=bio_chem_renames,
    index=bio_chem_renames,
    inplace=True)
print(bio_chem_corr_matrix)
```

```python
# Análisis Multivariado
# Matriz de Correlación para Datos Socio-demográficos

from numpy._core.defchararray import index

coexisting_factors = ["age", "sex", "anaemia", "diabetes", "high_blood_pressure", "smoking", "death_event"]
correlation_matrix = preprocessed_heart_failure[coexisting_factors].corr()
correlation_matrix.rename(
    columns={'age':'Edad','sex':'Género', 'anaemia':'Anemia','diabetes':'Diabetes','high_blood_pressure':'Presión alta','smoking':'Fumador','death_event':'Muerte'},
    index={'age':'Edad','sex':'Género', 'anaemia':'Anemia','diabetes':'Diabetes','high_blood_pressure':'Presión alta','smoking':'Fumador','death_event':'Muerte'},
    inplace=True)
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Crear la figura y los subplots (1 fila, 2 columnas)
fig, axs = plt.subplots(1, 2, figsize=(14, 6));
plt.suptitle('Evidencia de Valores Atípicos en Creatinina Fosfoquinasa y Creatinina en Suero', fontsize=16);

# Boxplot para Creatinina Fosfoquinasa
sns.boxplot(y=heart_failure['creatinine_phosphokinase'], ax=axs[0], color=selected_color);
axs[0].set_title('Creatinina Fosfoquinasa');
axs[0].set_ylabel('Creatinina Fosfoquinasa (mcg/L)');

# Calcular los límites del IQR para creatinine_phosphokinase
Q1_cpk = heart_failure['creatinine_phosphokinase'].quantile(0.25)
Q3_cpk = heart_failure['creatinine_phosphokinase'].quantile(0.75)
IQR_cpk = Q3_cpk - Q1_cpk
lower_bound_cpk = Q1_cpk - 1.5 * IQR_cpk
upper_bound_cpk = Q3_cpk + 1.5 * IQR_cpk

# Añadir líneas para los límites del IQR
axs[0].axhline(lower_bound_cpk, color='green', linestyle='--', label=f'Límite Inferior (1.5 IQR): {lower_bound_cpk:.2f}')
axs[0].axhline(upper_bound_cpk, color='red', linestyle='--', label=f'Límite Superior (1.5 IQR): {upper_bound_cpk:.2f}')
axs[0].legend(fontsize=8)

# Boxplot para Creatinina en Suero
sns.boxplot(y=heart_failure['serum_creatinine'], ax=axs[1], color=selected_color);
axs[1].set_title('Creatinina en Suero');
axs[1].set_ylabel('Creatinina en Suero (mg/dL)');

# Calcular los límites del IQR para serum_creatinine
Q1_sc = heart_failure['serum_creatinine'].quantile(0.25)
Q3_sc = heart_failure['serum_creatinine'].quantile(0.75)
IQR_sc = Q3_sc - Q1_sc
lower_bound_sc = Q1_sc - 1.5 * IQR_sc
upper_bound_sc = Q3_sc + 1.5 * IQR_sc

# Añadir líneas para los límites del IQR
axs[1].axhline(lower_bound_sc, color='green', linestyle='--', label=f'Límite Inferior (1.5 IQR): {lower_bound_sc:.2f}')
axs[1].axhline(upper_bound_sc, color='red', linestyle='--', label=f'Límite Superior (1.5 IQR): {upper_bound_sc:.2f}')
axs[1].legend(fontsize=8)

plt.tight_layout(rect=[0, 0.03, 1, 0.95]); # Ajustar el layout para evitar superposición del título
plt.show();
```

En resumen, la gráfica proporciona una visión clara de la prevalencia de cada característica sociodemográfica, destacando que estas comorbilidades afectan, de forma independiente, a mas del 30% de la población.

```python
# Análisis Univariado
# Gráficos de Barras
# Variables Socio-Demográficas (omitiendo edad y género)

categorical_features = ['anaemia', 'diabetes', 'high_blood_pressure', 'smoking'] # Se omitió 'sex'

# Diccionario para mapear nombres de variables a español
spanish_names = {
    'anaemia': 'Anemia',
    'diabetes': 'Diabetes',
    'high_blood_pressure': 'Presión Arterial Alta',
    'sex': 'Género',
    'smoking': 'Tabaquismo',
}

# Preparar los datos para el formato de pila 100%
plot_data = []
for col_name in categorical_features:
    counts = preprocessed_heart_failure[col_name].value_counts(normalize=True) * 100
    row_data = {'Variable': spanish_names[col_name]}
    # Asegurarse de que ambas categorías 0 y 1 estén presentes, incluso si una tiene 0% de ocurrencia
    row_data['Percentage_0'] = counts.get(0, 0)
    row_data['Percentage_1'] = counts.get(1, 0)
    plot_data.append(row_data)

stacked_percentages_df = pd.DataFrame(plot_data)

# Definir el orden para el eje y (variables) para una graficación consistente
y_axis_order_spanish = [spanish_names[col] for col in reversed(categorical_features)]
stacked_percentages_df['Variable'] = pd.Categorical(stacked_percentages_df['Variable'], categories=y_axis_order_spanish, ordered=True)
stacked_percentages_df = stacked_percentages_df.sort_values('Variable')
stacked_percentages_df = stacked_percentages_df.set_index('Variable')

# Colores para los dos estados (0 y 1)
# Percentage_0 (No / Mujer) será rosa
# Percentage_1 (Sí / Hombre) será azul
category_colors_map = {
    'Percentage_0': '#FFB6C1',  # Rosa
    'Percentage_1': '#ADD8E6'   # Azul
}

# Crear el gráfico de barras horizontales apiladas al 100%
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar Percentage_1 primero para que 'Sí / Hombre' quede a la izquierda, luego Percentage_0
stacked_percentages_df[['Percentage_1', 'Percentage_0']].plot(
    kind='barh', stacked=True, ax=ax,
    color=[category_colors_map['Percentage_1'], category_colors_map['Percentage_0']] # Asignar colores en el orden de las columnas graficadas
)

ax.set_title('Variables Sociodemográficas', fontsize=16)
ax.set_xlabel('Porcentaje (%)')
ax.set_ylabel('Variable Categórica')
ax.set_xlim(0, 100) # Asegurar que el eje x vaya de 0 a 100 para los porcentajes

# Personalizar las etiquetas de la leyenda - ahora reflejando el nuevo orden (Percentage_1 luego Percentage_0)
new_legend_labels = ['Sí', 'No'] # Etiquetas generales para 1 y 0
ax.legend(title='Categoría', labels=new_legend_labels)

ax.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
