#Práctica #9: Distribución de Datos con plt.subplots()

## Objetivo
Utilizar plt.subplots() para visualizar diferentes distribuciones generadas con NumPy y almacenadas en un DataFrame.

## Instrucciones:
- Genera un DataFrame con Pandas que contenga 1000 valores en cuatro columnas (Normal, Uniforme, Exponencial, Poisson).

- Crea una figura con plt.subplots(2, 2) para organizar los histogramas en una cuadrícula de 2x2.

- Grafica cada distribución en un subplot usando ax[i, j].hist().

- Agrega títulos a cada subplot y ajusta la disposición con tight_layout().

Sube un archivo txt que contenga el enlace a tu notebook.

## Preparación del entorno

```python
import numpy as np
import pandas as pd
# matplotlib es la biblioteca
# pyplot es un submódulo
# plt es el alias
import matplotlib.pyplot as plt
```

##Creación de las series de datos

```python
rng = np.random.default_rng(seed=42)

serie_normal = rng.normal(loc=0,scale=1,size=1000)
serie_uniforme = rng.uniform(low=0,high=100,size=1000)
serie_exponencial = rng.exponential(scale=100,size=1000)
serie_poisson = rng.poisson(lam=100,size=1000)
```

##Creación del dataframe

```python
data = pd.DataFrame({
    'Normal': serie_normal,
    'Uniforme': serie_uniforme,
    'Exponencial': serie_exponencial,
    'Poisson': serie_poisson
})
data
```

##Graficación
Se crea la cuadrícula de histogramas conforme a las instrucciones, con el uso de las librerías, Numpy,(NumPy team, 2026) Pandas, (Pandas team, 2026) y PyPlot (Hunter et al., 2026) de MatPlotLib.

```python
# using the variable axs for multiple Axes
fig, axs = plt.subplots(2, 2);
axs[0,0].set_title('Normal');
axs[0, 0].hist(data['Normal'], bins=30, color='skyblue', edgecolor='black', alpha=0.7);
axs[0, 0].set_xlabel('Valor')
axs[0, 0].set_ylabel('Frecuencia')

axs[0, 1].set_title('Uniforme');
axs[0, 1].hist(data['Uniforme'], bins=30, color='skyblue', edgecolor='black', alpha=0.7);
axs[0, 1].set_xlabel('Valor')
axs[0, 1].set_ylabel('Frecuencia')

axs[1, 0].set_title('Exponencial');
axs[1, 0].hist(data['Exponencial'], bins=30, color='skyblue', edgecolor='black', alpha=0.7);
axs[1, 0].set_xlabel('Valor')
axs[1, 0].set_ylabel('Frecuencia')

axs[1, 1].set_title('Poisson');
axs[1, 1].hist(data['Poisson'], bins=30, color='skyblue', edgecolor='black', alpha=0.7);
axs[1, 1].set_xlabel('Valor')
axs[1, 1].set_ylabel('Frecuencia')
# El prefijo 'r' es vital para que Python no interprete mal los backslashes (\)
plt.suptitle(r'Muestras de Distribución de Datos', fontsize=16)
fig.tight_layout()
```
