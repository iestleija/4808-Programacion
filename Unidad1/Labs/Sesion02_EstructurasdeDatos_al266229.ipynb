### **String, Str, Cadena: Tipo de Datos Primitivo**

Los tipos de datos String en Python son un arreglo de caracteres que forman cadenas para formar un mensaje o oración generalmente. Se pueden crear usando  comillas simples, dobles o triples.

La diferencia en cómo se pueden crear no es semántica, sino práctica:

Comillas simples ' ' y dobles " "
* Son equivalentes.
* Se usan para cadenas de una sola línea.
* Facilitan escapar comillas internas: indicarle al lenguaje que una comilla forma parte del texto y no marca el fin de la cadena.

Comillas triples ''' ''' o """ """
* Permiten cadenas multilínea.
* Conservan saltos de línea y formato.
* Se usan comúnmente para docstrings.

El tipo de dato es el mismo (str); solo cambia cómo se escribe y para qué se usa.


````python

profesor = "Sergio Castaño Giraldo"
web = 'Control Automático Educación'
cursos = """
1. Python
2. Matlab
3. MicroPython
4. Arduino
5. PIC
6. Control de Procesos
"""




print(profesor)
print(web)
print(cursos)

#Aquí \" indica que la comilla doble es contenido, no cierre de la cadena.
texto = "Juan dijo: \"hola\""

print(texto)

#Alternativa sin escapar

texto = 'Juan dijo: "hola"'

print(texto)
````

### **Tuplas (Tuple)**

Colecciones ordenadas e inmutables de longitud fija que, una vez asignada, no se puede modificar. La forma más sencilla de crearla es con una secuencia de valores separados por comas y entre paréntesis. Asimismo, contienen una secuencia heterogénea de elementos.



```python
In [1]: tup0 = ('foo', [1, 2], True)
Out[1]: ('foo', [1, 2], True)

In [2]: tup = (4, 5, 6)

In [3]: tup
Out[3]: (4, 5, 6)
```

Uso típico:

- Datos que no deben cambiar

- Registros fijos

````python
tup0 = ('foo', [1, 2], True)

print(tup0)

tup = (4, 5, 6)

tup
````

En muchos contextos, los paréntesis se pueden omitir, por lo que aquí también podríamos haber escrito:



``` python
#Este proceso se llama empaquetado
tup = 4, 5, 6

tup

(4, 5, 6)
```




````python
tup = 4, 5, 6

tup
````

Puedes convertir cualquier secuencia o iterador en una tupla invocando `tuple`:


```python
In [6]: tuple([4, 0, 2])
Out[6]: (4, 0, 2)

In [7]: tup = tuple('string')

In [8]: tup
Out[8]: ('s', 't', 'r', 'i', 'n', 'g')
```


````python

# Esta es una lista
print(tuple([4, 0, 2]))

#Esta es una cadena
tup = tuple('string')

print(tup)
````

Un problema particular es la construcción de tuplas que contengan 0 o 1 ítem: la sintaxis presenta algunas peculiaridades para estos casos. Las tuplas vacías se construyen mediante un par de paréntesis vacío; una tupla con un ítem se construye poniendo una coma a continuación del valor (no alcanza con encerrar un único valor entre paréntesis). Feo, pero efectivo. Por ejemplo:



```python
#Una tupla vacía

#Empaquetando una tupla con un elemento

len(singleton)
1
```



````python
# Una tupla vacía
empty = ()
#Empaquetando una tupla con un elemento
singleton = 'hello',    # <-- note trailing comma
print(len(empty))

print(len(singleton))
````

Se puede acceder a los elementos mediante corchetes, `[]`como ocurre con la mayoría de los demás tipos de secuencias. Al igual que en C, C++, Java y muchos otros lenguajes, las secuencias se indexan a **0** en Python.

Regla general:

Tupla sencilla: `estructura[indice]`

Tupla anidada: `estructura[indice_externo][indice_interno]`



```python
In [9]: tup[0]
Out[9]: 's'
```



````python
print(tup)
tup[0]
````

Al definir tuplas dentro de expresiones más complicadas, a menudo es necesario encerrar los valores entre paréntesis, como en este ejemplo de creación de una tupla de tuplas:



```python
In [10]: nested_tup = (4, 5, 6), (7, 8)

In [11]: nested_tup
```

````python
nested_tup = (4, 5, 6), (7, 8)

print(nested_tup)

#Acceder a la primera tupla
print(nested_tup[0])

#Acceder a los elementos de la tupla interna (primera)

print(nested_tup[0][0])
print(nested_tup[0][1])

print(nested_tup[1])
````

**Métodos de Tupla**

Las tuplas tienen pocos métodos porque son inmutables.

````python
#1. count()

#Cuenta número de veces que aparece un elemento.

t = (1, 2, 2, 3, 2)
t.count(2)
````

### **Listas**

Colecciones ordenadas y mutables. Que una lista sea *ordenada* significa que los elementos mantienen el orden en el que fueron insertados y ese orden no cambia automáticamente.

Que una lista sea *mutable* significa que puede modificarse después de haber sido creada, sin necesidad de crear una nueva lista.

Uso típico:
- Datos secuenciales

- Conjuntos de observaciones

- Variables explicativas

En analítica: almacenar mediciones, series temporales, columnas simples.

Se pueden definir mediante corchetes `[]` o con la función `list` función.




### **Diccionarios**

El diccionario dictes quizás la estructura de datos integrada más importante de Python. En otros lenguajes de programación, los diccionarios se denominan mapas hash o matrices asociativas .

Un diccionario almacena una colección de pares **clave-valor** , donde clave y valor son objetos de Python. Cada clave está asociada a un valor para que este pueda recuperarse, insertarse, modificarse o eliminarse fácilmente dada una clave específica.

 Una forma de crear un diccionario es usar llaves `{}`y dos puntos para separar claves y valores:



```python
In [83]: empty_dict = {}

In [84]: d1 = {"a": "some value", "b": [1, 2, 3, 4]}

In [85]: d1
Out[85]: {'a': 'some value', 'b': [1, 2, 3, 4]}
```



````python
empty_dict = {}

d1 = {"a": "some value", "b": [1, 2, 3, 4]}

d1
````

Puede acceder, insertar o configurar elementos utilizando la misma sintaxis que para acceder a los elementos de una lista o tupla:



```python
In [86]: d1[7] = "an integer"

In [87]: d1
Out[87]: {'a': 'some value', 'b': [1, 2, 3, 4], 7: 'an integer'}

In [88]: d1["b"]
Out[88]: [1, 2, 3, 4]
```



````python
print(d1)
d1[7] = "an integer"

print(d1)

print(d1["b"])
````

Puede comprobar si un diccionario contiene una clave utilizando la misma sintaxis utilizada para comprobar si una lista o tupla contiene un valor:



```python
In [89]: "b" in d1
Out[89]: True
```



El método `keys` y `values`proporciona iteradores de las claves y valores del diccionario, respectivamente. El orden de las claves depende del orden de inserción, y estas funciones generan las claves y valores en el mismo orden:



```python
In [99]: list(d1.keys())
Out[99]: ['a', 'b', 7]

In [100]: list(d1.values())
Out[100]: ['some value', [1, 2, 3, 4], 'an integer']
```



Si necesita iterar sobre las claves y los valores, puede utilizar el itemsmétodo para iterar sobre las claves y los valores como 2-tuplas:



```python
In [101]: list(d1.items())
Out[101]: [('a', 'some value'), ('b', [1, 2, 3, 4]), (7, 'an integer')]
```



### **Set - Conjuntos**

Un conjunto es una colección desordenada de elementos únicos. Se puede crear de dos maneras: mediante la función `set` o mediante un literal de conjunto entre llaves:



```python
In [124]: set([2, 2, 2, 1, 3, 3])
Out[124]: {1, 2, 3}

In [125]: {2, 2, 2, 1, 3, 3}
Out[125]: {1, 2, 3}
```



````python
set([2, 2, 2, 1, 3, 3])

print(set([2, 2, 2, 1, 3, 3]))
print({2, 2, 2, 1, 3, 3})
````

Los conjuntos admiten operaciones matemáticas como unión, intersección, diferencia y diferencia simétrica. Considere estos dos conjuntos de ejemplo:

````python
a = {1, 2, 3, 4, 5}

b = {3, 4, 5, 6, 7, 8}

print(a.union(b))

print(a | b)
````

La intersección contiene los elementos presentes en ambos conjuntos. Se puede utilizar el operador `&`o el método: `intersection`

````python
a.intersection(b)

a & b
````
