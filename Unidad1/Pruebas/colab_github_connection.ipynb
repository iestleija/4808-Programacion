##Programación para analítica descriptiva y predictiva - Herramientas en la nube.
Para la materia de Programación para analítica descriptiva y predictiva es importante contar con un entorno ágil y colaborativo. Para ello se requiere un entorno estable y común para la ejecución de los segmentos de código en Python.
Se requiere de la misma forma un mecanismo de almacenamiento tanto de los archivos de práctica como de los datos que serán procesados durante las mencionadas prácticas.

* Colab presenta un entorno de prácticas y documentación integrado en una sola
interface, cuenta con un entorno virtual de ejecución. De la mima manera tiene una interface que permite mezclar código, los resultados de la ejecución del mismo y texto formateado para documentación.
* Google drive por su parte proporciona espacio de almacenamiento en la nube con la funcionalidad completa de un sistema de archivos.
* gitHub adiciona el control de versiones para controlar el desarrollo de cualquier proyecto.

Colab permite la integración con Google Drive y gitHub dándole al usuario la posibilidad de tomar la mayor ventaja para ejecutar, documentar, almacenar y colaborar en las prácticas relativas a la materia de este curso.
