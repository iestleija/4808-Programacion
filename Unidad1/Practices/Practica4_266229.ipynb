## Crea dos matrices de 3x3 con números aleatorios entre 1 y 10

### Preparación del ambiente
Se importa numpy y se establece una semilla global para los calculos aleatorios.

```python
import numpy as np

# Set the global seed
np.random.seed(12)
```

### Creación de las dos matrices
Se crean las matrices A y B, ambas con dimension 3x3 y valores aleatorios entre 1 y 10.

```python
# create matrices
A = np.random.randint(1, 11, size=(3, 3))
B = np.random.randint(1, 11, size=(3, 3))
print(A)
print(B)
```

### 1. Suma Matricial
Debido a que las matrices fueron generadas por numpy, la adición de las mismas se realiza conforme al álgebra lineal, elemento a elemento. *Es importante anotar que las listas nativas de Python durante la suma se concatenan.*

```python
# Matrices addition executed by Numpy
print(A+B)
```

### 2. Resta Matricial
De igual forma que la suma la sustracción matricial la ejecuta Numpy, acorde a la convención matemática.

```python
# Sustraction of matrices made by Numpy
print(A-B)
```

### 3. Multiplicación de Matriz por un Escalar
Se presenta ahora la multiplicación de la matriz A por un escalar, la que conforme a la teoría matemática resulta de multiplicar el escalar por cada elemento y ese valor colocarlo en la misma posición del elemento en la nueva matriz.

```python
# multiplication of a matrix by an scalar value
print(A*2)
```

### 4. Multiplicación algebráica de matrices
La multiplicación de matrices se efectúa con el método np.dot de Numpy, el cual sigue el principio de calcular cada nuevo elemento de la matriz resultante mediante el producto punto de la fila de la primera matriz y la columna de la segunda que intersectan en la posición de ese elemento precisamente.

```python
# Algebraic matrices multiplication made with np.dot method
print(np.dot(A, B))
```
