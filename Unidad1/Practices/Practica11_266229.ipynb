## Carga del conjunto de datos Penguins

```python
penguins = sns.load_dataset("penguins")
penguins
```

## Graficación con subplots y Seaborn

```python
import matplotlib.ticker as ticker

# Crear la figura y los subplots (2 filas, 2 columnas)
fig, ax = plt.subplots(2, 2, figsize=(10, 6));
plt.suptitle(r'Dataset Penguins Visualizado con Seaborn + PyPlot.subplots', fontsize=16)

# Crear subplots automáticos con `FacetGrid`
ax[0,0].set_title("Histograma de longitudes de aletas");
ax[0,0].set_xlabel("Largo de aletas (mm)");
ax[0,0].set_ylabel("Frecuencia");
sns.histplot(data=penguins, x="flipper_length_mm", bins=30, color="skyblue", kde=True, ax=ax[0, 0]);

ax[0,1].set_title("Boxplot de masa corporal por especie");
ax[0,1].set_xlabel("Especie");
ax[0,1].set_ylabel("Masa corporal (kg)");
kilo_formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1000:.0f}')
ax[0,1].yaxis.set_major_formatter(kilo_formatter)
sns.boxplot(data=penguins, x="species", y="body_mass_g", color="skyblue", linecolor="#137", linewidth=.75, ax=ax[0,1]);

ax[1,0].set_title("Dispersión entre masa corporal y largo de aletas");
ax[1,0].set_xlabel("Largo de aletas (mm)");
ax[1,0].set_ylabel("Masa corporal (kg)");
ax[1,0].yaxis.set_major_formatter(kilo_formatter)
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", color="0.3", ax=ax[1,0]);

ax[1,1].set_title("Violin plot de masa corporal según la especie");
ax[1,1].set_xlabel("Especie");
ax[1,1].set_ylabel("Masa corporal (kg)");
ax[1,1].yaxis.set_major_formatter(kilo_formatter)
ax[1,1].grid(True, axis="y", linestyle="--")
sns.violinplot(data=penguins, x="species", y="body_mass_g", color="0.3", fill=False);
plt.tight_layout()
```
