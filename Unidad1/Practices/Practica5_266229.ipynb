### Instrucciones
Uso de máscaras booleanas.

Crea un arreglo unidimensional con 15 números aleatorios entre 1 y 50. Luego:

1. Selecciona los elementos mayores que 25 utilizando una máscara booleana.
2. Cuenta cuántos elementos son mayores que 25.

## Preparación del ambiente
Se importa la librería Numpy y se establece la semilla para valores aleatorios para Numpy.

```python
import numpy as np
np.random.seed(12)
```

## Creación del arreglo unidimensional con 15 números aleatorios entre 1 y 50.

```python
# array creation with a dimension of 15 and random values between 1 and 50
array_a = np.random.randint(1,51,15)
print(array_a)
```

### 1. Identificación y despliegue de los elementos mayores a 25
Se crea una máscara boleana para filtrar los elementos cuyo valor es mayor a 25, se guardan esos elementos y se despliegan en pantalla.

```python
mask = array_a > 25 # boolean mask is defined to filter those elements having a value over 25
over_25 = array_a[mask] # Boolean mask appplied to recover the matching elements.
print("Valores del arreglo mayores a 25:",over_25)
```

### 2. Conteo de los elementos mayores a 25
Se despliega el conteo de los valores mayores a 25 encontrados en el arreglo inicial.

```python
print(len(over_25),"elementos son mayores a 25")
```
