##File opening
Now the data source file *informe_acciones.txt* is opened and its handler assigned to the *sample_shares* variable. Notice it is in a nested folder called Classfiles.
*There was a configuration mismatch referring to the file encoding that was resolved with the use of chardet.*

````python
import chardet

with open('/content/drive/MyDrive/ClassFiles/informe_acciones.txt', 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))

detected_encoding = result['encoding']
print(f"Detected encoding: {detected_encoding}")

sample_shares = open('/content/drive/MyDrive/ClassFiles/informe_acciones.txt','r',encoding = detected_encoding)
````

##Data Processing
The logic to process this file is as follows:
```
Initialize the rows list
Initialize the data row with null (None) values.
For every row in the file:
  get the share code if exists
  get the owner company if exists
  get the share value if exists
    As value marks the completion of a data row,
      Save the data row in the row list and,
      Reset the data row values.
```



###Functions definition
As we have several repetitive tasks, they are coded in the following functions.

````python
import re
def get_code(linea):
  if re.search('^Asunto:', linea):
    return re.search(r'[ ][-][ ](.*)$', linea).group(1)
  return None

def get_company(linea):
  if re.search('^Te escribo', linea):
    return re.search(r'acción de (.*)[(]', linea).group(1)
  return None

def get_price(linea):
  if re.search('^Valor actual:', linea):
    return re.search(r'[:][ ](.*)$', linea).group(1)
  return None

def add_share(code,company,value):
  shares_list.append([code,company,value])
````

###Initialize main variables

````python
shares_list = list()
code = None
company = None
price = None
contador = 0
````

###Main file processing loop
This is the code implementing the data processing logic as stated before

````python
for share in sample_shares:
  share=share.rstrip()

  # Get the data row values only once until share price is found
  if code == None:
    code = get_code(share)
    if code != None:
      continue

  if company == None:
    company = get_company(share)
    if code != None:
      continue

  if price == None:
    price = get_price(share)
    # When price is found, the data row is complete, added to the list and reset.
    if price != None:
      add_share(code,company,price)
      code = None
      company = None
      price = None
      contador = contador + 1
    continue

shares_list.append(["","Shares in the report",contador])

print(shares_list)
````
