# Creación de un arreglo unidimensional de 20 elementos con valores aleatorios entre 1 y 100

```python
# Crea un arreglo unidimensional de 20 elementos con valores aleatorios entre 1 y 100
array = np.random.randint(1,101,20)
print(array)
```

### 1. Obtención del valor máximo del arreglo

```python
print(f'El valor máximo del arreglo es: {np.max(array)}')
```

### 2. Obtención del valor mínimo del arreglo

```python
print(f'El valor mínimo del arreglo es: {np.min(array)}')
```

### 3. Obtención de la media del arreglo

```python
print(f'El valor del promedio del arreglo es: {np.mean(array)}')
```

### 4. Arreglo adicionado con el escalar 10
Nótese que en esta operación numpy simula realizar una suma con otro arreglo equidimensional al primero cuyos elementos todos son igual al escalar 10.

```python
print(f'El arreglo original sumado a un escalar de 10 es: {array + 10}')
```
