## Creación del dataframe

```python
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul','Ago','Sep','Oct','Nov','Dic']
data = pd.DataFrame({
    'Mes': meses,
    'Ventas': np.random.randint(2000, 5000, size=12),
    'Gastos': np.random.randint(1000, 4000, size=12)
})
data
```

## Creación de la plantilla de gráficos con GridSpec
Como parte de la librería matplotlib,(Hunter et al., 2026) GridSpec permite organizar la distribución de gráficos en una cuadrícula flexible, proporcionando la capacidad de extender los gráficos horizontal o verticalmente tomando mas de una celda.

```python

fig = plt.figure(figsize=(6, 6))  # Crear figura de 6 * 6 pulgadas
gs = gridspec.GridSpec(2, 2)  # Definir una cuadrícula de 2x2

# Crear subplots en posiciones específicas
ax1 = fig.add_subplot(gs[0, :])  # Primer subplot (fila completa)
ax2 = fig.add_subplot(gs[1, 0])  # Segundo subplot (fila 1, columna 0)
ax3 = fig.add_subplot(gs[1, 1])  # Tercer subplot (fila 1, columna 1)

x = np.arange(len(meses))
ancho_barra = 0.35
ax1.bar(x - ancho_barra/2, data['Ventas'], ancho_barra, label='Ventas', color='skyblue')
ax1.bar(x + ancho_barra/2, data['Gastos'], ancho_barra, label='Gastos', color='green')
ax1.set_xticks(x)
ax1.set_xticklabels(data['Mes'])
ax1.set_title('Comparativo de Ventas vs Gastos');
ax1.legend()
fig.tight_layout()

ax2.plot(data['Mes'], data['Ventas'], marker='o', linestyle='-', color='skyblue', label='Ventas')
ax2.plot(data['Mes'], data['Gastos'], marker='s', linestyle='--', color='green', label='Gastos')
ax2.tick_params(axis='x', labelrotation=45)

ax3.scatter(data['Mes'], data['Ventas'], color='skyblue', s=100, alpha=0.7, edgecolors='black')
ax3.scatter(data['Mes'], data['Gastos'], color='green', s=100, alpha=0.7, edgecolors='black')
ax3.tick_params(axis='x', labelrotation=45)
```
